```python
from diffusers import StableDiffusionPipeline
import torch
import gradio as gr
```

#Model setup

```python
model_id = "prompthero/openjourney"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

## Test prompt

```python
prompt = "retro serie of different cars with different colors and shapes, mdjrny-v4 style"
```

```python
image = pipe(prompt).images[0]
image.save("./retro_cars.png")
```

#Gradio interface

## Define function and input/output for interface

```python
def Text2Image(query):
    image = pipe(query + ', mdjrny-v4 style').images[0]
    return image
```

```python
inputs = gr.inputs.Textbox(lines=3, label="Enter a sentence to generate an image:")
outputs = gr.outputs.Image(type="pil", label="Generated Image")
```

##Defining Gradio interface

```python
interface = gr.Interface(fn=Text2Image, inputs=inputs, outputs=outputs, 
                          title="Text2Image using StableDiffusionPipeline", 
                          description="Generate an image based on the given input text. In order to save the image, you can simply right-click and save it to your local system.", 
                          allow_download=True, 
                          interface_size=(800,600),
                         examples=[['Banana'],['A fit model'],['A cat driving a car'],['Big mumbajumbas'],['ur mom']],
                         examples_per_page=5).launch()
```
